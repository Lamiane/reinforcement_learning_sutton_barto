```python
% matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

size = 100
t = xrange(size)
logt = np.log(t)
t_a_chosen = sorted(np.random.randint(0, size, 10))
nt = [0]*size
for i in t_a_chosen:
    for idx in range(i, size):
        nt[idx]+=1
        

logt_nt = logt/nt
sqrt_l_n = np.sqrt(logt_nt)

plt.figure(figsize=(10,7.5))
plt.plot(logt, c='k', label='logarithm over t', lw=2)
plt.plot(np.array(nt)/3., c='b', label='n_t - number of times action was choosen until now (/3)', lw=2)
plt.plot(logt_nt, c='g', label='log/n_t', lw=2)
plt.plot(sqrt_l_n, c='r', label='sqrt(log(t)/n_t)', lw=2)
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
           ncol=2, mode="expand", borderaxespad=0.)

for xc in t_a_chosen:
    plt.axvline(x=xc, color='k', linestyle='--')
plt.show()
```
